```python
import torch
from diffusers import StableDiffusionPipeline
import gradio as gr
from PIL import Image
import datetime
import os
```

```python
# Load Stable Diffusion pipeline using CPU
pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float32  # Use float32 for CPU
)

pipe = pipe.to("cpu")  # Send to CPU
```

```python
def generate_logo(prompt, steps, guidance):
    image = pipe(prompt, num_inference_steps=steps, guidance_scale=guidance).images[0]

    # Save the image
    timestamp = datetime.datetime.now().strftime("%Y%m%d%H%M%S")
    file_path = f"/tmp/logo_{timestamp}.png"
    image.save(file_path)

    return image, file_path
```

```python
with gr.Blocks() as demo:
    gr.Markdown("## 🎨 BrandGenAI")

    with gr.Row():
        with gr.Column():
            prompt = gr.Textbox(label="Enter your logo idea", placeholder="e.g., Futuristic tech startup logo with glowing AI brain")
            steps = gr.Slider(10, 50, value=25, step=1, label="Steps (Quality vs Time)")
            guidance = gr.Slider(5, 15, value=7.5, step=0.5, label="Creativity (Guidance Scale)")
            generate_btn = gr.Button("🚀 Generate Logo")

        with gr.Column():
            logo_output = gr.Image(label="Generated Logo")
            download_link = gr.File(label="Download Logo")

    def handle_generate(prompt, steps, guidance):
        return generate_logo(prompt, steps, guidance)

    generate_btn.click(handle_generate, inputs=[prompt, steps, guidance], outputs=[logo_output, download_link])

demo.launch(share=True)
```
